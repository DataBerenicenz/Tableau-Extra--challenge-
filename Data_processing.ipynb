```python
import os, glob
import pandas as pd
```

```python
path = "Resources/"

all_files = glob.glob(os.path.join(path,"JC-2020*.csv"))

all_csv = (pd.read_csv(f, sep= ",") for f in all_files)
citi_bike_merged = pd.concat(all_csv, ignore_index= True)
```

```python
citi_bike_merged = citi_bike_merged.drop(columns=['bikeid'], axis=1)
citi_bike_merged
```

```python
citi_bike_merged = citi_bike_merged.sort_values(by=['starttime'])
citi_bike_merged
```

```python
conversions= {
    'gender': {
        1: 'Male',
        2: 'Female',
        0: 'Null'
    }}
```

```python
citi_bike_merged = citi_bike_merged.copy()
for k,v in conversions.items():
    citi_bike_merged[k] = citi_bike_merged[k].replace(v)
```

```python
citi_bike_merged
```

```python
citi_bike_merged = citi_bike_merged[citi_bike_merged.gender != 'Null']
```

```python
citi_bike_merged.to_csv("citi_bike_merged.csv")
```
